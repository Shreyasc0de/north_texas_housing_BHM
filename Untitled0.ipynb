```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seed for reproducibility
np.random.seed(42)

# Define Counties and their "True" baseline price per sq ft (Intercepts)
# Based on May 2026 trends
counties = ['Dallas', 'Tarrant', 'Collin', 'Denton']
true_intercepts = [185, 170, 210, 195]  # Base price per sq ft
n_houses_per_county = 250

data = []
for i, county in enumerate(counties):
    # Generate random square footage (1200 to 4500 sq ft)
    sqft = np.random.normal(2500, 700, n_houses_per_county)
    sqft = np.clip(sqft, 1000, 6000)

    # Calculate price with some random noise (error)
    # Price = (Base County Rate * sqft) + random noise
    noise = np.random.normal(0, 15000, n_houses_per_county)
    price = (true_intercepts[i] * sqft) + noise

    for s, p in zip(sqft, price):
        data.append({'county': county, 'sqft': s, 'price': p})

df = pd.DataFrame(data)

# Step 2: Preprocessing for PyMC
# We need to map county names to integers (0, 1, 2, 3)
df['county_idx'] = pd.Categorical(df['county']).codes

print(df.head())
```

```python
import pymc as pm

# Coordinate system for the model
county_idxs = df.county_idx.values
coords = {"county": counties, "obs": np.arange(len(df))}

with pm.Model(coords=coords) as hierarchical_model:
    # 1. Hyper-priors (The "Global" Texas Level)
    mu_alpha = pm.Normal("mu_alpha", mu=190, sigma=50)
    sigma_alpha = pm.Exponential("sigma_alpha", 1.0)

    # 2. County-level priors (The "Hierarchical" intercepts)
    # Each county gets its own alpha, but they "share" the hyper-priors
    alpha = pm.Normal("alpha", mu=mu_alpha, sigma=sigma_alpha, dims="county")

    # 3. Model for the Slope (shared across counties for now)
    beta = pm.Normal("beta", mu=0, sigma=10) # Effect of sqft variation

    # 4. Expected Value (Linear Regression equation)
    # We use the county_idxs to tell PyMC which alpha belongs to which row
    mu = alpha[county_idxs] * df.sqft.values / 1000 # Scaling for math stability

    # 5. Likelihood (The observation)
    sigma = pm.Exponential("sigma", 1.0)
    price_obs = pm.Normal("price_obs", mu=mu, sigma=sigma, observed=df.price.values / 1000)

    # 6. Sampling (The "MCMC" Magic)
    trace = pm.sample(1000, tune=1000, target_accept=0.9)
```

```python
import arviz as az

# 1. Visualize the Posterior Distributions and Sampler Chains
az.plot_trace(trace, var_names=["mu_alpha", "alpha", "beta"])
plt.tight_layout()
```

```python
# 2. Get a summary table of the results
summary = az.summary(trace, var_names=["alpha", "beta"], round_to=2)
print(summary)
```

```python
# 3. Forest Plot to compare counties
az.plot_forest(trace, var_names=["alpha"], combined=True)
plt.title("Estimated Base Price per County (Hierarchical Intercepts)")
plt.grid(True)
```

```python
# 4. Check if the model can recreate the original data
with hierarchical_model:
    pm.sample_posterior_predictive(trace, extend_inferencedata=True)

az.plot_ppc(trace, num_pp_samples=100)
```
